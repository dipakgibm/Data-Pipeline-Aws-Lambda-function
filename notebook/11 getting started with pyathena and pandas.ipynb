```python
pd.read_sql(
    ''' 
    select 
        case when org is null  then 'IND'
        ELSE 'ORG' end,
        count(*) as acivity_count
        from ghact.parquets group by case when org is null then 'IND'
        else 'ORG'
        end
        limit 10
        
    ''',conn
)
```
